```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
print(boston.keys())
```

```python
print(boston.DESCR)
```

```python
import pandas as pd
df = pd.DataFrame(boston.data)
df.columns = boston.feature_names
df['MEDV'] = boston.target
df.head(6)
```

```python
y=np.array(df['MEDV'])
X=np.array(df['LSTAT'])
```

```python
import numpy.polynomial.polynomial as poly
```

```python
print(y)
print(X)
```

```python
plt.scatter(X,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
d = 1
beta_hat = poly.polyfit(X,y,d)
beta = np.array([1,0.5,0,2])
# Plot true and estimated function
xp = np.linspace(0,40,100)
yp = poly.polyval(xp,beta)
yp_hat = poly.polyval(xp,beta_hat)
plt.ylim(0,50)

plt.plot(xp,yp_hat,'g-',linewidth=2)

# Plot data
plt.scatter(X,y)
plt.legend(['d=1', 'Data'], loc='upper left')
plt.grid()
plt.xlabel('X')
plt.ylabel('y')
```

```python
d = 15
beta_hat = poly.polyfit(X,y,d)
# Plot true and estimated function
xp = np.linspace(0,40,100)
yp = poly.polyval(xp,beta)
yp_hat = poly.polyval(xp,beta_hat)
plt.ylim(0,50)

plt.plot(xp,yp_hat,'r-',linewidth=2)

# Plot data
plt.scatter(X,y)
plt.legend(['d=15', 'Data'], loc='upper left')
plt.grid()
plt.xlabel('X')
plt.ylabel('y')
```

```python
import  sklearn.model_selection
```

```python
k = 5
kfo = sklearn.model_selection.KFold(n_splits=k,shuffle=True)

# Model orders to be tested
dtest = np.arange(0,10)
nd = len(dtest)

RSSts = np.zeros((nd,k))

# Loop over the folds
for isplit, Ind in enumerate(kfo.split(X)): # enumerate returns a counter (isplit) and a list (Ind)
        
    # Get the training data in the split
    Itr, Its = Ind 
    #kfo.split( ) produced Ind, which contains a pair of index sets for each fold 
    xtr = X[Itr]
    ytr = y[Itr]
    xts = X[Its]
    yts = y[Its]
    
    # Loop over the model order
    for it, d in enumerate(dtest):
    
        # Fit data on training data
        beta_hat = poly.polyfit(xtr,ytr,d)
        
        # Measure RSS on test data
        yhat = poly.polyval(xts,beta_hat)
        RSSts[it,isplit] = np.mean((yhat-yts)**2)
```

```python
RSS_mean = np.mean(RSSts,axis=1) #note mean is taken over the second dimension, i.e. all folds for the same model order
RSS_se = np.std(RSSts,axis=1,ddof=1)/np.sqrt(k)
plt.errorbar(dtest, RSS_mean, yerr=RSS_se, fmt='-')
plt.ylim()
plt.xlabel('Model order')
plt.ylabel('mean Test RSS')
plt.grid()
```

```python
imin = np.argmin(RSS_mean)
print("The selected model order is {0:d}".format(dtest[imin]))
```

```python
# Find the minimum RSS target
imin = np.argmin(RSS_mean)
RSS_tgt = RSS_mean[imin] + RSS_se[imin]

# Find the lowest model order below the target
I = np.where(RSS_mean <= RSS_tgt)[0]
iopt = I[0]
dopt = dtest[iopt]

plt.errorbar(dtest, RSS_mean, yerr=RSS_se, fmt='-')

# Plot the line at the RSS target
plt.plot([dtest[0],dtest[imin]], [RSS_tgt, RSS_tgt], '--')

# Plot the line at the optimal model order
plt.plot([dopt,dopt], [0,500], 'g--')

plt.ylim(20,40)
plt.xlim(0,10)
plt.xlabel('Model order')
plt.ylabel('mean Test RSS')
plt.grid()

# Print results
print("The estimated model order is %d" % dopt)
```

```python
beta_min=poly.polyfit(X,y,dtest[imin])
yp_min=poly.polyval(xp,beta_min)
beta_ose=poly.polyfit(X,y,dtest[iopt])
yp_opt=poly.polyval(xp,beta_ose)
plt.plot(xp,yp_min,'r-')
plt.plot(xp,yp_opt,'g-')

plt.scatter(X,y)
plt.legend(['Rss_min','Rss_ose','data'],loc='upper right')
plt.ylim(0,40)
plt.xlim(0,40)
plt.grid()
plt.show()
```
